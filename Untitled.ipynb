```python
from autoscraper import AutoScraper
```

```python
amazon_url = "https://www.amazon.in/s?k=Iphone"

wanted_list = ["₹34,999","Apple iPhone XR (64GB) - (Product) RED","5,159"]
```

```python
scraper = AutoScraper()
result = scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_b15b':'Title','rule_rcyq':'Price','rule_gjmp':'Rating'})
scraper.keep_rules(['rule_b15b','rule_rcyq','rule_gjmp'])
scraper.save('amazon-search')
```

```python
Results = scraper.get_result_similar("https://www.amazon.in/s?k=Samsung",group_by_alias=True)
```
